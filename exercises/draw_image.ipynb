```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
plt.imshow(cv2.rectangle(blank_img, pt1=(300,0), pt2=(510,150), color=(0, 255, 0), thickness=8))
```

```python
plt.imshow(cv2.rectangle(blank_img,pt1=(200,200), pt2=(300,300), color=(255,0,0), thickness=5))
```

```python
plt.imshow(cv2.circle(img=blank_img, center=(100,100), radius=5, color=(0,0,255), thickness=5))
```

```python
plt.imshow(cv2.line(img=blank_img, pt1=(0,0), pt2=(500,500), color=(255,0,0), thickness=2))
```

```python
plt.imshow(cv2.circle(img=blank_img, center=(100,400), radius=20, color=(0,0,255), thickness=-1))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
plt.imshow(cv2.putText(img=blank_img, text='test', org=(100, 350), color=(0,255,0), fontFace=font, fontScale=3, thickness=5))
```
