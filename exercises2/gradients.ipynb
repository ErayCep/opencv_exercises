```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('../DATA/sudoku.jpg', 0)
```

```python
display_image(img)
```

```python
sobelx = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=-1)
```

```python
display_image(sobelx)
```

```python
sobely = cv2.Sobel(img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=-1)
display_image(sobely)
```

```python
laplace = cv2.Laplacian(img, ddepth=cv2.CV_64F)
```

```python
display_image(laplace)
```

```python
blended = cv2.addWeighted(src1=sobelx, alpha=0.5, src2=sobely, beta=0.5, gamma=0)
display_image(blended)
```
