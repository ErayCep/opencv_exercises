```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../images/twitter_KkJ91P32_400x400.jpg')
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
sword = fix_img[:,50:200]
sword_gray = cv2.cvtColor(sword, cv2.COLOR_RGB2GRAY)
ret2, thresh2 = cv2.threshold(sword_gray, 100, 255, cv2.THRESH_TRUNC)
```

```python
ax, plots = plt.subplots(1, 2)
plots[0].imshow(sword)
plots[1].imshow(thresh2, cmap='gray')
```
