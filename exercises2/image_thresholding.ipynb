```python
import cv2
import matplotlib.pyplot as plt
from matplotlib import image
%matplotlib inline
```

```python
img = cv2.imread('../images/twitter_KkJ91P32_400x400.jpg', 0)

fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img, cmap='gray')
```

```python
sword = img[:,50:200]
ret2, thresh2 = cv2.threshold(sword, 200, 255, cv2.THRESH_TRUNC)
```

```python
ax, plots = plt.subplots(1, 2)
plots[0].imshow(sword, vmin=0, vmax=255)
plots[1].imshow(thresh2, cmap='gray', vmin=0, vmax=255)
```

```python
img_gray = cv2.cvtColor(fix_img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
```

```python
ret, thresh1 = cv2.threshold(img_gray, 100, 255, cv2.THRESH_TRUNC)
```

```python
plt.imshow(thresh1, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
ret, thr1 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
show_pic(thr1)
```
