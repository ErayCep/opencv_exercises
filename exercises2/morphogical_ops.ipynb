```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def load_image():
    blank_img = np.zeros(shape=(600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img, text='text', org=(50,300), fontFace=font, fontScale=8, color=(255,255,255), thickness=10)
    return blank_img
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_image()
display_image(img)
```

```python
kernel = np.ones(shape=(5,5), dtype=np.uint8)
kernel
```

```python
erode_img = cv2.erode(img, kernel=kernel, iterations=1)
display_image(erode_img)
```

```python
img = load_image()
```

```python
white_noise = np.random.randint(low=0, high=2, size=(600,600))
white_noise
```

```python
white_noise = white_noise * 255
```

```python
noise_img = white_noise + img
display_image(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN, kernel=kernel)
display_image(opening)
```

```python
black_noise = np.random.randint(low=0, high=2, size=(600, 600))
black_noise = black_noise * -255
```

```python
black_noise_img = black_noise + img
```

```python
black_noise_img[black_noise_img == -255] = 0
```

```python
display_image(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE, kernel=kernel, iterations=1)
display_image(closing)
```
